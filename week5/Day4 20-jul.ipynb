##### joint plot
* combination of two plots.By default it have scatter plot,histogram
* sns.jointplot()

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.jointplot(data=iris)
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,color='g')
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='hex',color='orange')
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='kde',color='r')
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='reg',color='m')
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='resid',color='b')
```

##### pair plot
* it displays multiple plots at a time in a single graph
* pairwise relationship
* By default it return scatter plot
* sns.pairplot()

```python
sns.pairplot(iris)
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris,hue='species')
```

```python
#draw a pairplot with different colours
sns.pairplot(iris,hue='species',palette='husl')
```

```python
sns.pairplot(iris,hue='species',palette='dark')
```

```python
sns.pairplot(iris,hue='species',markers=['o','D','s'])
```

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width'])
```

##### heatmap
* it is used to find correlation between two columns in data

```python
help(sns.heatmap)
```

```python
sns.load_dataset('tips')
```

```python
t=sns.load_dataset('tips')
```

```python
t.head()
```

```python
t.corr()
```

```python
sns.heatmap(t.corr())
```

```python
sns.heatmap(t.corr(),annot=True)
```

```python
sns.heatmap(t.corr(),center=0,linewidth=5,linecolor='g')
```
