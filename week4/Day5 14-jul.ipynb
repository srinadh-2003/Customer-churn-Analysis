#### Data imputation(used to deal with missing values)
* It is a method for retaining the mojority of dataset's set and information by substituting missing data with a different value.
* methods 
 - using mean value
 - using median value
 - using mode
 - filling missing values with a constant

```python
from sklearn.impute import SimpleImputer
```

```python
import pandas as pd
import numpy as np
```

```python
d1={
    "a":pd.Series([10,89,76,np.nan,56],index=[1,2,3,4,5]),
    "g":pd.Series([454,34,23,34],index=[1,3,4,5]),
    "c":pd.Series([13,45,35,35],index=[1,2,3,4])
}
df=pd.DataFrame(d1)
df
```

* fit_transform() from sklearn,preprocessing class is used to preprocessing data for model training

```python
si=SimpleImputer(strategy='median')
si.fit_transform(df)
```

```python
df.median()
```

```python
si=SimpleImputer(strategy='mean')
si.fit_transform(df)
```

```python
df.mean()
```

```python
si=SimpleImputer(strategy='most_frequent')
si.fit_transform(df)
```

```python
si=SimpleImputer(strategy='constant',fill_value=-1)
si.fit_transform(df)
```

* Feature scaling
  * it is atechnique that involves the values of features or variables in a dataset into a similar scale.
- it is the process of normalizing the range of features in a dataset
- real world data often contain features that are varying in degrees of magnitude,range and units
- Therefore inorder for machinelearning models to interpret these features on same scale,we need to perform scaling

* Scaling techniques
  - used to convert raw data into meaningful data
    - standardizing data
    - data range
    - normalizing data
    - robust scaling

* standardizing data(column wise/label wise/feature wise)
  - standard format refers to data that has 0 mean and unit variance(standard deviation=1),the process of converting  data into this format is called data standardization
