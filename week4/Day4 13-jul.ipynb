### Data cleaning
* isnull() : to check the data having null values or not.
* notnull() : to get the not null values
* dropna() : to remove the null values
* fillna() : to fill the null values with meaningfull data.
* replace() : to replace the data
* NaN : not a number


```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Salary_Data.csv')
df
```

```python
# isnull() : to check the data having null values or not.
df.isnull()
```

```python
# to print the number of null values
df.isnull().sum()
```

```python
# to print only null value records
df[df.isnull().any(axis=1)]
```

```python
# notnull()
df.notnull()
```

```python
# to  count not null records
df.notnull().sum()
```

```python
# to print notnull records
df[df.notnull().all(axis=1)]
```

```python
# to remove null records
df.dropna()
```

```python
#fillna(): to fill null values with meaning full data
df.fillna('missing_value',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df1=pd.read_csv('Salary_Data.csv')
df1
```

```python
df2=pd.read_csv('Salary_Data.csv')
df2
```

```python
# replace():to replace the data
df.replace(to_replace='missing_value',value='NaN',inplace=True)
df
```

```python
df
```

```python
# based on specific column name
df['Gender'].replace(to_replace='Male',value='M',inplace=True)
df
```

```python
df['Salary'].replace(to_replace=[6000,6500],value=[1000,2000],inplace=True)
df
```
